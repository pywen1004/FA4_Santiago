# 1. Write a Python program to print the even numbers from a given list.

```python
List = []
while True:
    a = int(input("Enter a number or press 0 to finish the list: "))
    if a == 0:
        break
    List.append(a)

for n in List:
    if n % 2 == 0:
        print(n, end=" ")
```

# 2. Generate a Python list of all the even numbers between 10 to 60.

```python
a = int(input("Enter a number for the starting range: "))
b = int(input("Enter a number for the ending range: "))
even_numbers = [i for i in range(a, b) if i % 2 == 0]
print(even_numbers)
```

# Write a program to create a function show_feu_professor() using the following conditions:

* It should accept the professorâ€™s name and salary and display both.
* If the salary is missing in the function call then assign default value 70000 to salary.

```python
def show_feu_professor(name, salary=70000):
    print(name, "\t\tPhp ", salary)

professors_details = []

for i in range(5):
    name = input("Enter name: ")
    salary = input("Enter salary (press Enter for default value): ")
    if salary:
        professors_details.append((name, int(salary)))
    else:
        professors_details.append((name,))

print("\nProfessor Details:")
for professor_detail in professors_details:
    if len(professor_detail) == 2:
        show_feu_professor(professor_detail[0], professor_detail[1])
    else:
        show_feu_professor(professor_detail[0])
```
